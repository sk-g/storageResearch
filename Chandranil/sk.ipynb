```python
import os
import pandas as pd
import gc
import collections
```

```python
#os.listdir(os.path.join(os.getcwd(),'All_in_one'))
def loader(fname):
    data = pd.read_csv(fname,
                      index_col = False,
                      engine = 'python')
    return data
```

```python
data = loader((os.path.join(os.getcwd(),'All_in_one','output.csv')))
```

```python
data.drop(data.index[[0]],inplace = True)
data.head()
```

```python
def change_process(dataframe):
    l = len(dataframe)
    for i in range(1,len(dataframe)):
        #drawProgressBar(i/l)
        name = dataframe.ix[i,'ProcessName']
        name = list(name)
        name = collections.deque(name)
        for _ in range(len(name)-1,-1,-1):
            x = name.pop()
            if x == '(':
                break;
        name = ''.join([_ for _ in name])
        
        dataframe.at[i,'ProcessName'] = name
    #print(dataframe.head(15))
    return dataframe
data = change_process(data)
data.head()
```

```python
data.reset_index(inplace = True)
data.head()
```
